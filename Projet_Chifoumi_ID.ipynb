```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# --- PARTIE 1 : DATA AUGMENTATION ---
# G√©n√©rateur d'images avec modifications al√©atoires
train_datagen = ImageDataGenerator(
    rescale=1./255,       # Normalisation (0-1)
    rotation_range=20,    # Rotation l√©g√®re
    width_shift_range=0.2, # D√©calage horizontal
    height_shift_range=0.2, # D√©calage vertical
    shear_range=0.2,      # D√©formation
    zoom_range=0.2,       # Zoom
    horizontal_flip=True, # Miroir
    fill_mode='nearest'
)

# Chargement depuis les dossiers
# Attention : Assure-toi que le chemin '.' correspond bien l√† o√π sont tes dossiers Pierre/Feuille/Ciseaux
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/Projet_Chifoumi',                  # Dossier racine (ici le dossier courant)
    classes=['Pierre', 'Feuille', 'Ciseaux'],
    target_size=(150, 150), # Taille standardis√©e
    batch_size=32,
    class_mode='categorical'
)

# --- PARTIE 2 : CR√âATION DU MOD√àLE CNN ---
model = tf.keras.models.Sequential([
    # Convolution 1 : D√©tecter les formes simples
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2,2),

    # Convolution 2 : D√©tecter les formes complexes
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Convolution 3 : Affiner les d√©tails
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    # Classification
    tf.keras.layers.Flatten(),            # Mise √† plat
    tf.keras.layers.Dropout(0.5),         # Pour √©viter le "par c≈ìur" (Overfitting)
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax') # 3 sorties (Pierre, Feuille, Ciseaux)
])

# Compilation du mod√®le
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# --- PARTIE 3 : LANCEMENT DE L'ENTRA√éNEMENT ---
# Lancement (Epochs = nombre de fois o√π l'IA voit tout le dataset)
history = model.fit(train_generator, epochs=100, steps_per_epoch=30)

model.save("/content/drive/MyDrive/Projet_Chifoumi/mon_ia.h5")
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

# 1. On r√©cup√®re les noms des classes (pour savoir qui est 0, 1 et 2)
# Cela permet d'avoir la correspondance chiffr√©e -> nom
class_names = {v: k for k, v in train_generator.class_indices.items()}
print("L'IA connait ces classes :", class_names)

# 2. On prend une nouvelle photo
print("üì∏ Fais un geste (Pierre, Feuille ou Ciseaux)...")
filename = take_photo('test_chifoumi.jpg')

# 3. On pr√©pare l'image comme pour l'entra√Ænement (150x150 et normalisation)
img = image.load_img(filename, target_size=(150, 150))
x = image.img_to_array(img)
x = x / 255.0  # N'oublie pas la normalisation !
x = np.expand_dims(x, axis=0) # On ajoute une dimension car l'IA attend un "paquet" d'images

# 4. Pr√©diction
predictions = model.predict(x)
score = predictions[0]

# 5. R√©sultat
print("\n--- R√âSULTATS ---")
print(f"Probabilit√©s brutes : {score}")
gagnant_index = np.argmax(score) # On prend l'index du score le plus √©lev√©
gagnant_nom = class_names[gagnant_index]
confiance = score[gagnant_index] * 100

print(f"ü§ñ L'IA pense que c'est : **{gagnant_nom}**")
print(f"Certitude : {confiance:.2f}%")
```

```python
import numpy as np
import random
from tensorflow.keras.preprocessing import image

# Dictionnaire pour traduire les chiffres en mots
classes = {0: 'Pierre', 1: 'Feuille', 2: 'Ciseaux'}

def jouer_une_manche():
  pts_ordi = 0
  pts_joueur = 0

  while pts_ordi < 3 and pts_joueur < 3:

    print("\n" + "="*30)
    print("ü§ñ L'ORDINATEUR vs THE GOAT")
    print("="*30)

    # 1. L'ordinateur fait son choix
    choix_ordi = random.choice(['Pierre', 'Feuille', 'Ciseaux'])
    print("ü§î L'ordinateur a choisi son geste... √† toi !")

    # 2. Prise de la photo
    print("üì∏ Fais ton geste et clique sur Capture...")
    filename = take_photo('manche_actuelle.jpg')

    # 3. Pr√©paration de l'image pour le mod√®le
    img = image.load_img(filename, target_size=(150, 150))
    x = image.img_to_array(img)
    x = x / 255.0  # Normalisation (tr√®s important !)
    x = np.expand_dims(x, axis=0)

    # 4. Pr√©diction
    predictions = model.predict(x, verbose=0)
    index_gagnant = np.argmax(predictions)
    confiance = np.max(predictions)

    geste_joueur = classes[index_gagnant]

    # 5. R√©sultats
    print(f"\nüß† L'IA voit : **{geste_joueur}** (S√ªre √† {confiance*100:.1f}%)")
    print(f"üíª L'Ordi avait choisi : **{choix_ordi}**")
    print("-" * 30)

    # 6. Qui a gagn√© ?

    if confiance*100 <= 80 :
      print("Refais ton signe !")
    elif geste_joueur == choix_ordi:
        print("ü§ù √âGALIT√â !")
    elif (geste_joueur == 'Pierre' and choix_ordi == 'Ciseaux') or \
         (geste_joueur == 'Feuille' and choix_ordi == 'Pierre') or \
         (geste_joueur == 'Ciseaux' and choix_ordi == 'Feuille'):
        pts_joueur += 1
        print(f"üèÜ Tu remportes la manche ! \n Utilisateur {pts_joueur} - {pts_ordi} Ordinateur !")
    else:
        pts_ordi += 1
        print(f"üíÄ L'ordinateur gagne cette manche. \n Utilisateur {pts_joueur} - {pts_ordi} Ordinateur !")


  if pts_ordi == 3:
    print(f"üíÄ L'ordinateur √† gagner la partie {pts_ordi} - {pts_joueur}.")
  else:
    print(f"üèÜ Tu as gagn√© la partie {pts_joueur} - {pts_ordi}.")

# --- LANCER LE JEU ---
# Tu peux relancer cette ligne autant de fois que tu veux


jouer_une_manche()
```
